Python群集型態
1. List(清單): [30, 20, 40]
    - 放同類型的東西
2. Dict(字典): {"姓名":"Elwing", "身高":175}
    - 拿來聚合成一個複雜的資料
前兩個最重要, 因為這兩個加起來就是JSON格式
3. Set(不重複的集合): {"周", "王", "何"}
    - 全台灣所有的姓氏集合(Set)
    - 所有大於3小於10的整數(Set)
    - 是不會讓你自訂key的(沒有key)
    - 不重複, 所以你寫幾次都只會幫你保留一個
    - 因為沒有key, 沒有所謂順序
4. Tuple(簡化版本的字典): ()
    - 一個pixel: {"R":255, "G":255, "B":255}
    - 對於我們來說, rgb有共識的, 簡化一下 (255, 255, 255)
    - 你可以把他想成 {0:255, 1:255, 2:255}
    - tuple的key是0 1 2
    - !!! tuple可以當成字典的key(因為key必須是永不可變的東西)
    - !!! 字典key必須是不可變

```python
# 第一個練習:
# 之前的是 a / b 後的十位小數
# extra 1. 整除就停下來
# extra 2. 如果循環停下來

a, b = 20, 7
# 來解決a > b的情況
result = str(a // b) + "."
a = a % b

# 空的set不能用{}, 這會變成空的字典
# 養成好習慣 list() set() dict() 來創空的
seen = set()
while True:
    # 2 -> 20
    a10 = a * 10
    # 每當看到一個新的數字, 就加進去
    if a10 in seen:
        print("循環了")
        break
    else:
        seen.add(a10)
    # 20 // 7
    result = result + str(a10 // b)
    # 20 % 7 -> 6(a)
    a = a10 % b
    if a == 0:
        break
print(result)
```

```python
import random

LOW, HIGH = 1, 49
def get_ticket(low, high, count=6):
    # 買彩券
    lottery = set()
    while len(lottery) < count:
        n = random.randint(low, high)
        lottery.add(n)
        # print("!!!", n, lottery)
    return lottery

prize = get_ticket(LOW, HIGH)
# 處理特別號
# 迴圈, 隨機, 直到產生不是在六個裡面
special = None
while True:
    n = random.randint(LOW, HIGH)
    if not n in prize:
        special = n
        break

print("頭彩:", prize)
print("特別號是:", special)


count = 0
SCORE_WANT = 10
PRICE = 50
while True:
    # 買了一張
    count = count + 1
    # 兌獎
    # 分數: 正號2 特別號1
    # 正號的確認
    score = 0

    # 幫我寫一個迴圈, 一直一直買, 直到三獎以上才停下來, 算一下你花了多少錢(50)
    buy = get_ticket(LOW, HIGH)
    # print("我買的:", buy)
    bingo = prize & buy
    score = score + 2 * len(bingo)
    # print("對中的正號:", bingo)
    # print("目前分數:", score)
    # 特別號的確認
    if special in buy:
        score = score + 1
        # print(special, "特別號中了")
    # print("目前分數:", score)
    if score >= SCORE_WANT:
        print("買了幾張:", count, "花了:", count * PRICE)
        print("買的是:", buy, "正號:", bingo, "特別號:", special in buy)
        break
```
